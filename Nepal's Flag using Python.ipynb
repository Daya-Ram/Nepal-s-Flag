```python
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_nepal_flag():
    fig, ax = plt.subplots(figsize=(10, 12))  # Adjusted figsize to match intended output size

    # Define the main colors
    red_color = '#DC143C'
    blue_color = '#0033A0'
    white_color = '#FFFFFF'

    # Draw the flag outline
    # First triangle (bottom)
    triangle1 = patches.Polygon([(0, 0), (10, 0), (0, 10)], closed=True, edgecolor=blue_color, facecolor=red_color, linewidth=10)
    ax.add_patch(triangle1)
    
    # Second triangle (top)
    triangle2 = patches.Polygon([(0, 10), (8, 8), (0, 18)], closed=True, edgecolor=blue_color, facecolor=red_color, linewidth=10)
    ax.add_patch(triangle2)

    # Draw the moon (upper part)
    moon_outer = patches.Circle((2, 12), 1.2, color=white_color, zorder=1)
    moon_inner = patches.Circle((2, 13.0), 1.1, color=red_color, zorder=2)
    ax.add_patch(moon_outer)
    ax.add_patch(moon_inner)
    
    # Draw the sun (lower part)
    sun_outer = patches.Circle((2, 3), 1.0, color=white_color, zorder=1)
    sun_inner = patches.Circle((2, 3), 0.8, color=red_color, zorder=2)
    ax.add_patch(sun_outer)
    ax.add_patch(sun_inner)

    # Draw the rays of the sun
    for i in range(12):
        angle = i * 30
        ray = patches.RegularPolygon((2, 3), numVertices=3, radius=1.2, orientation=angle * (3.14159 / 180), color=white_color, zorder=1)
        ax.add_patch(ray)

    # Set limits and remove axes
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 18)
    ax.axis('on')  # Remove axes
    
    # Save the image to desktop using os module
    desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop', 'nepal_flag.png')
    plt.savefig(desktop_path)

    plt.show()

draw_nepal_flag()
```
